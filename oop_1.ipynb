```python
class Atm:

  #constructor
  def __init__(self):
    self.pin =''
    self.balance=0
    # print("constructor got executed !")
    self.menu()


  def menu(self):

    while(True):
      user_input=input("""
      Hi how can I help You ?

      1.Press 1 to create Account
      2.Press 2 to withdrawl Money
      3.Press 3 to change pin
      4.Press 4 to check balance
      5.Anything else to exit
      """)

      if user_input=='1':
        self.create_account()

      elif user_input=='2':
        self.withdrawl_money()

      elif user_input=='3':
        self.change_pin()

      elif user_input=='4':
        self.check_balance()

      else:
        # exit()
        break



  #create pin /account

  def create_account(self):
    user_pin=input("Enter Your Pin To Create Account :")
    self.pin=user_pin
    print("Pin created Successfully !!!")
    user_balance=int(input("Enter Your balance : "))
    self.balance=user_balance
    print("Account crated successfully !!!")

  #withdrawl money

  def withdrawl_money(self):
    user_amount=int(input("Enter Amount to withdrawl :"))
    if user_amount<=self.balance:
      self.balance=self.balance - user_amount
      print("Amount withdrawl successfull !!!")

    else:
      print("Insufficient Balance to withdrawl !!!")


  #chnage pin
  def change_pin(self):
    old_pin=input("Enter Your old pin :")

    if old_pin==self.pin:
      new_pin=input("Enter Your new Pin :")
      self.pin=new_pin
      print("Pin changed successfully !!!")
    else:
      print("Please Enter Correct Old Pin !!")


  #check balance
  def check_balance(self):
    user_pin=input("Enter Your pin : ")

    if user_pin==self.pin:
      print("Available Balance is :",self.balance)

    else:
      print("Wrong pin Entered Try Again !!!")



```

## Class Diagram

- A **Class Diagram** is a visual representation of classes, their attributes, methods, and the relationships between them.
- It is mainly used in UML (Unified Modeling Language) to design and plan the structure of a system.
- To draw a class diagram:  
  - Create a rectangle divided into three parts:  
    - Top section for the **class name**  
    - Middle section for **attributes** (variables)  
    - Bottom section for **methods** (functions)  
  - Use arrows or lines to show relationships like inheritance or association between classes.

# Magic Methods

## Magic Methods in Python

- **Magic Methods** (also called **dunder methods**) are special methods in Python that start and end with double underscores (e.g., `__init__`, `__str__`, `__len__`).
- They are used to define the behavior of custom objects with built-in Python operations like printing, addition, comparison, etc.

### Common Examples:
- `__init__(self)`  Constructor, used to initialize object properties.
- `__str__(self)`  Returns a string representation of the object (used in `print()`).
- `__len__(self)`  Returns the length using `len()` function.
- `__add__(self, other)`  Defines behavior for the `+` operator.

# Lets Code our own datatype - Fraction

```python
3/4 *1/2  # instead of 0.375 we want 3/8
```

```python
class Fraction:

  #parameterized constructor
  def __init__(self,x,y):
    self.num=x
    self.den=y

  def __str__(self):## whatever we try to print object that will get executed according to this magic method __str__
    return '{}/{}'.format(self.num,self.den)

  def __add__(self,other): # gets call we try to add object
    new_num=self.num*other.den +other.num*self.den
    new_den=self.den*other.den

    return '{}/{}'.format(new_num,new_den)


  def __sub__(self,other): # gets call we try to add object
    new_num=self.num*other.den - other.num*self.den
    new_den=self.den*other.den

    return '{}/{}'.format(new_num,new_den)

  def __mul__(self,other): # gets call we try to add object
    new_num=self.num*other.num
    new_den=self.den*other.den

    return '{}/{}'.format(new_num,new_den)


  def __truediv__(self,other): # gets call we try to add object
    new_num=self.num*other.den
    new_den=self.den*other.num

    return '{}/{}'.format(new_num,new_den)
```

```python
fr1=Fraction(3,4)
fr2=Fraction(4,5)
print(fr1)
print("add :",fr1+fr2)
print("sub :",fr1-fr2)
print("multiplication :",fr1*fr2)
print("division :",fr1/fr2)
```
